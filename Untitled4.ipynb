```python
def abl(x,y,i):
    d=x[i]**2+y[i]**2
    n=sum([a**2+b**2 for a,b in zip(x,y)]) + (1-sum(x))**2 + (sum(y))**2
    return d/n
alpha=[0.3,0.5]
counter=0
epsilon=1e-4
N_points=30
probsA = np.empty(N_points**4)
probsB = np.empty(N_points**4)
for x0 in np.linspace(0+epsilon,1-epsilon,N_points):
    for x1 in np.linspace(0+epsilon,1-epsilon,N_points):
        for y0 in np.linspace(0+epsilon,1-epsilon,N_points):
            for y1 in np.linspace(0+epsilon,1-epsilon,N_points):
                probsA[counter]=abl([x0,x1],[y0,y1],0)
                probsB[counter]=abl([x0,x1],[y0,y1],0)
                counter+=1

probsA=probsA.reshape(N_points,N_points,N_points,N_points)
probsB=probsB.reshape(N_points,N_points,N_points,N_points)
```
